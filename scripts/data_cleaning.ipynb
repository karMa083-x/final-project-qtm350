```python
import pandas as pd
import requests

region_codes = ["EAS", "SAS"]

all_countries = []

for region in region_codes:
    url = f"https://api.worldbank.org/v2/region/{region}/country?format=json&per_page=500"
    response = requests.get(url)
    data = response.json()
    df_region = pd.DataFrame(data[1])
    all_countries.append(df_region)

countries_df = pd.concat(all_countries, ignore_index=True)
print(countries_df[['id', 'iso2Code', 'name']].head(30))
```

```python
asian_countries_df = countries_df[
    countries_df['region'].apply(lambda x: x['id'] in region_codes)
].copy()

asian_countries_df = asian_countries_df[
    asian_countries_df['id'].str.len() == 3
].copy()

print(asian_countries_df[['id', 'name']].head(20))
print(asian_countries_df.shape)
```

```python
country_codes = asian_countries_df['id'].tolist()
countries = ";".join(country_codes)

print(countries[:200])   
print(len(country_codes))  # how many countries
```

```python
# Life expectancy

indicator = "SP.DYN.LE00.IN"

url = f"https://api.worldbank.org/v2/country/{countries}/indicator/{indicator}?format=json&per_page=20000"

response = requests.get(url)
life_raw = response.json()

life_df_raw = pd.DataFrame(life_raw[1])
life_df_raw.head(20)
```

```python
# Under-5 mortality
indicator = "SH.DYN.MORT"
url = f"https://api.worldbank.org/v2/country/{countries}/indicator/{indicator}?format=json&per_page=20000"

response = requests.get(url)
mort_raw = response.json()
mort_df_raw = pd.DataFrame(mort_raw[1])

mort_df_raw.head()
```

```python
# Adolescent fertility
indicator = "SP.ADO.TFRT"
url = f"https://api.worldbank.org/v2/country/{countries}/indicator/{indicator}?format=json&per_page=20000"

response = requests.get(url)
fert_raw = response.json()
fert_df_raw = pd.DataFrame(fert_raw[1])

fert_df_raw.head()
```

```python
life = life_df_raw.copy()

life = life[['countryiso3code', 'date', 'value']]
life.columns = ['country_code', 'year', 'life_expectancy']

life['year'] = pd.to_numeric(life['year'], errors='coerce')
life['life_expectancy'] = pd.to_numeric(life['life_expectancy'], errors='coerce')

life = life[(life['year'] >= 1990) & (life['year'] <= 2024)]

life.head()
```

```python
mort = mort_df_raw.copy()

mort = mort[['countryiso3code', 'date', 'value']]
mort.columns = ['country_code', 'year', 'mortality_u5']

mort['year'] = pd.to_numeric(mort['year'], errors='coerce')
mort['mortality_u5'] = pd.to_numeric(mort['mortality_u5'], errors='coerce')

mort = mort[(mort['year'] >= 1990) & (mort['year'] <= 2024)]

mort.head()
```

```python
fert = fert_df_raw.copy()

fert = fert[['countryiso3code', 'date', 'value']]
fert.columns = ['country_code', 'year', 'fertility_adolescent']

fert['year'] = pd.to_numeric(fert['year'], errors='coerce')
fert['fertility_adolescent'] = pd.to_numeric(fert['fertility_adolescent'], errors='coerce')

fert = fert[(fert['year'] >= 1990) & (fert['year'] <= 2024)]

fert.head()
```

```python
# merge 
df = life.merge(mort, on=['country_code', 'year'], how='inner') \
         .merge(fert, on=['country_code', 'year'], how='inner')

df.head()
print(df.columns)
```

```python
# missing values
df = df.sort_values(['country_code', 'year']).copy()

num_cols = ['life_expectancy', 'mortality_u5', 'fertility_adolescent']

df[num_cols] = (
    df.groupby('country_code')[num_cols]
      .transform(lambda x: x.interpolate())
)
```

```python
df = df.dropna()
```

```python
# add country names
country_map = dict(zip(asian_countries_df['id'], asian_countries_df['name']))

df['country'] = df['country_code'].map(country_map)
df[['country_code', 'country']].drop_duplicates().head(20)
```

```python
# reorder columns
df = df[['country', 'country_code', 'year',
         'life_expectancy', 'mortality_u5', 'fertility_adolescent']]

df = df.sort_values(['country_code', 'year']).reset_index(drop=True)

df.head()
```

```python
df.to_csv("asian_population_clean.csv", index=False)
```
